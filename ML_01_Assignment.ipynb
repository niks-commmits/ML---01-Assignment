1. What is a parameter?
  -  A parameter is a value or a variable that is used to
     define or configure something within a function, method, or system. In different contexts, it can have slightly varied meanings, but the core idea is the same.

2. What is correlation?
   What does negative correlation mean?
  -  Correlation refers to a statistical relationship or
     association between two or more variables. When two variables are correlated, it means that changes in one variable are associated with changes in another. Correlation can describe the strength and direction of this relationship.

     Negative correlation: When one variable increases, the other variable decreases (or vice versa). For example, the relationship between the amount of time spent exercising and weight might show a negative correlation: more exercise often leads to weight loss.

3. Define Machine Learning. What are the main components in
   Machine Learning?
   - Machine Learning (ML) is a branch of artificial
     intelligence (AI) that focuses on creating algorithms and models that allow computers to learn from and make predictions or decisions based on data, without being explicitly programmed.

     The main Components in machine learning:

     Data (Training and Test Data)

     Model (The structure that represents the learned patterns)

     Algorithm (The process of learning from data)

     Features (Characteristics of the data)

     Training (Learning from data)

     Evaluation (Assessing model performance)

     Prediction (Making predictions on new data)

     Feedback Loop (Continuous improvement based on feedback)

4. How does loss value help in determining whether the model   
   is good or not?
 -   The loss value is a critical metric in machine  
     learning that helps determine how well or poorly a model is performing. It measures the difference between the model's predicted output and the actual target values (ground truth).

5. What are continuous and categorical variables?
  -  Continuous and categorical variables are two fundamental
     types of variables used in data analysis and machine learning. They describe different types of data and require different methods of analysis and modeling.

6. How do we handle categorical variables in Machine Learning?
   What are the common techniques?
  -  Handling categorical variables in machine learning is an
     essential step because most machine learning algorithms require numerical input, and categorical data cannot be used directly.
     common techniques are:-
     Label Encoding, One-Hot Encoding, Frequency Encoding, Target Encoding, Binary Encoding.

7. What do you mean by training and testing a dataset?
  -  Training and testing a dataset are key steps in building
     and evaluating a machine learning model. These terms refer to how data is used during the process of developing a model, helping to ensure that the model is effective and generalizes well to unseen data.

8. What is sklearn.preprocessing?
  -  sklearn.preprocessing is a module in the
     Scikit-learn library in Python that provides various functions to preprocess data before feeding it into machine learning algorithms. Preprocessing is an essential step in machine learning because the quality and format of the data can significantly impact the performance of the model.

9. What is a Test set?
  -  A test set is a subset of data used in machine learning to
     evaluate the performance of a trained model. After training a machine learning model using a training set, the test set is used to simulate how the model would perform on new, unseen data. It is an essential part of the model evaluation process to ensure the model generalizes well to new data and is not overfitting to the training data.

10. How do we split data for model fitting (training and
     testing) in Python?
   - In Python, particularly using Scikit-learn, We can
     easily split data into training and testing sets using the train_test_split function. This function is part of the sklearn.model_selection module and provides a simple way to split your data into two subsets for model training and evaluation.  

11. Why do we have to perform EDA before fitting a model to the
     data?
  -  Exploratory Data Analysis (EDA) is a crucial step in the
     machine learning workflow, and performing EDA before fitting a model to the data is important for several reasons. EDA helps you understand the structure and characteristics of your data, which can significantly influence how you preprocess, model, and evaluate it.

15.  What is causation? Explain difference between correlation
     and causation with an example.
  -   Causation refers to a relationship between two variables
      where one variable (the cause) directly affects the other variable (the effect).
     Correlation:
      Correlation measures the strength and direction of a relationship between two variables. It simply tells us that the variables are related in some way, but it does not imply any cause-and-effect relationship.
     Causation:
      Causation is when one variable directly influences another. If a change in variable A causes a change in variable B, we can say there is a causal relationship.

16. What is an Optimizer? What are different types of  
     optimizers? Explain each with an example.
  -  Optimizers play a crucial role in training machine  
     learning models by adjusting the weights and minimizing the loss function. Different optimizers like SGD, Adam, RMSprop, and Momentum have their unique features and are chosen based on the nature of the problem being solved.

17.  What is sklearn.linear_model ?
  -  The sklearn.linear_model module in scikit-learn provides  
     a set of linear models that can be used for both regression and classification tasks. Linear models are based on the concept of modeling the relationship between the input variables (features) and the output variable (target) using linear equations.

18.  What does model.fit() do? What arguments must be given?
  -  The model.fit() method in machine learning is used to  
     train the model on the provided dataset. It adjusts the model's internal parameters to learn patterns from the data, based on the provided input (features) and output (target) values.

19. What does model.predict() do? What arguments must be given?
  -  The model.predict() method in machine learning is used to
     make predictions based on the model that has already been trained using the fit() method. After training a model (using model.fit()), you use model.predict() to make predictions on new, unseen data.

20. What are continuous and categorical variables?
  -   Continuous Variables:
      Continuous variables are those that can take any value within a given range. They can have an infinite number of possible values, which can be decimals or fractions. These values can be measured, and the data can be ordered or ranked.
      Categorical Variables:
      Categorical variables are those that represent categories or groups. These variables have a finite number of distinct values, which may or may not have an inherent order.

21.  What is feature scaling? How does it help in Machine
     Learning?
  -  Feature scaling refers to the process of standardizing or
     normalizing the range of independent variables or features in a dataset. The goal of feature scaling is to adjust the scales of features so that they all have comparable magnitudes and units.

22. How do we perform scaling in Python?
  -  In Python, scaling of features can be easily performed
     using the scikit-learn library, which provides two main methods for scaling: Normalization (Min-Max Scaling) and Standardization (Z-score Scaling).

23. What is sklearn.preprocessing?
  -  sklearn.preprocessing is a module in scikit-learn, a  
     Python library for machine learning, that provides various tools for data preprocessing. Data preprocessing is a critical step in the machine learning pipeline as it helps prepare the raw data into a suitable format for modeling.

24.  How do we split data for model fitting (training and
     testing) in Python?
  -  In Python, we can split data into training and testing  
     sets using train_test_split from the sklearn.model_selection module. This function splits the dataset into two subsets: one for training the model and the other for evaluating its performance.

25.  Explain data encoding?
  -  Data encoding is a process used in machine learning to
     convert categorical data into a numerical format so that machine learning algorithms can process and analyze it.
